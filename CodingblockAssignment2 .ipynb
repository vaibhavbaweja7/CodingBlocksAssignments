```python
# Generating random numbers

# (a:0.5, b:0.2, c:0.3)

import numpy as np
import matplotlib.pyplot as plt
def random_func():
    r=np.random.random()
    
    if r<0.5:
        return "a"
    if r>=0.5 and r<0.8:
        return "b"
    return "c"
```

```python
N=1000
a,b,c =0,0,0
for ix in range(N):
    rn= random_func()
    
    if rn=="a":
         a+=1
if rn=="b":
         b+=1
if rn=="c":
         c+=1
print(a,b,c)       
```

```python
# Generate Normal distribution in 2D
import random
def gauss_2d(mu, sigma):
    x = random.gauss(mu, sigma)
    y = random.gauss(mu, sigma)
    return (x, y)
gauss_2d(0,1)
```

```python
# Convert to Standard Normal Distribution
import random
def gauss_2d(mu, sigma):
    x = random.gauss(mu, sigma)
    z = (x - mu)/sigma
    return z
gauss_2d(0,1)
```
